```python
import pandas as pd
import numpy as np
```

```python
df1= pd.read_excel("Data_Train.xlsx")
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.dropna(inplace=True)
```

```python
df1.shape
```

```python
df1["Destination"].value_counts() 
```

```python
#df1[df1["Destination"] == 'New Delhi'].Destination="Delhi"
df1['Destination'] = df1['Destination'].str.replace('New Delhi','Delhi')
```

```python
dict={"1 stop":1,"non-stop":0,"2 stops":2,"3 stops":3,"4 stops":4}
```

```python
df1["Total_Stops"]=df1["Total_Stops"].map(dict)
```

```python
df1.head()
```

```python
df1["Additional_Info"].value_counts()
```

```python
# dropping Additional_Info and Route

df1.drop(["Additional_Info","Route"], inplace=True, axis=1)
```

```python
df1.head()
```

```python
df1["DOJ_date"]=pd.DatetimeIndex(df1["Date_of_Journey"]).day
```

```python
df1["DOJ_month"]=pd.DatetimeIndex(df1["Date_of_Journey"]).month
df1["DOJ_year"]=pd.DatetimeIndex(df1["Date_of_Journey"]).year
```

```python
df1.head()
```

```python
df1["DOJ_year"].value_counts()
```

```python
# dropping DOJ year and DOJ

df1.drop(["DOJ_year","Date_of_Journey"], inplace=True, axis=1)
```

```python
df1.head()
```

```python
df1["Arrival_hour"]=pd.DatetimeIndex(df1["Arrival_Time"]).hour
df1["Arrival_min"]=pd.DatetimeIndex(df1["Arrival_Time"]).minute
```

```python
df1.head()
```

```python
# Working through the duration field


duration=list(df1["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:
        if "h" in duration[i]:
            duration[i]= duration[i].strip() + " 0m"
        else:
            duration[i]= "0h " + duration[i]
            
duration_hours=[]
duration_mins=[]

for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep='h')[0]))
    duration_mins.append(int(duration[i].split(sep='m')[0].split()[-1]))
```

```python
df1["Duration_hour"]=duration_hours
df1["Duration_min"]=duration_mins
```

```python
df1.head()
```

```python
df1["Dep_Hour"]=pd.DatetimeIndex(df1["Dep_Time"]).hour
df1["Dep_min"]=pd.DatetimeIndex(df1["Dep_Time"]).minute
```

```python
df1.drop(["Dep_Time","Arrival_Time","Duration"], axis=1, inplace=True)
```

```python
df1.head()
```

```python
airline=pd.get_dummies(df1["Airline"], drop_first=True)
source=pd.get_dummies(df1["Source"], drop_first=True)
destination=pd.get_dummies(df1["Destination"], drop_first=True)
```

```python
df1=pd.concat([df1, airline, source, destination], axis=1)
```

```python
df1.drop(["Airline","Source","Destination"], axis=1, inplace=True)
```

```python
df1.head()
```

```python
X=df1.copy()
```

```python
X.drop("Price", axis=1, inplace=True)
```

```python
X.head()
```

```python
y=df1["Price"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(18,18))

sns.heatmap(df1.corr(), annot=True, cmap="RdYlGn")
```

```python
# train test split

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
# linear regression feature importance
from sklearn.linear_model import LinearRegression

# define the model
model = LinearRegression()
# fit the model
model.fit(X_train, y_train)
# get importance
importance = model.coef_
```

```python
# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
```

```python
# plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```

```python
y_pred_lr=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred_lr
```

```python
print(r2_score(y_test,y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestRegressor

model_rf=RandomForestRegressor(n_estimators=500)
```

```python
model_rf.fit(X_train,y_train)
```

```python
y_pred_rf=model_rf.predict(X_test)
```

```python
print(r2_score(y_test,y_pred_rf))
```

```python
## Hyper Parameter Optimization

params={
 "n_estimators"    : [100,200,300, 400, 500, 600 ] ,
 "max_features"        : [ 'auto','sqrt','log2'],
  "max_depth"        : [ 3,5,10, None],   
 "min_samples_leaf"            : [ 1,2,3,4]   ,
    "criterion" : ["mse", "mae"]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
random_search=RandomizedSearchCV(model_rf,param_distributions=params,n_iter=5,n_jobs=-1,cv=5,verbose=3)
```

```python
random_search.fit(X_train,y_train)
```
